### Functions and If statements

```python
def has_student_passed(score):
  """
  This is a multiline comment.
  This function takes the score and returns pass if
  the score is above 40.
  """
  if score >= 40:
    print("The student has passed the course.") # Gets executed when if condition evaluates to true
  else:
    print("The student has not passed the course.") # Gets exectured when if conditon evaluates to false
  return
```

###  For Loop

```python
for s in scores:
  print(f"s={s},")
  has_student_passed(s)
```
